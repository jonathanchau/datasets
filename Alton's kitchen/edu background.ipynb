##convert all percentage numbers in table into floats


```python
immigration_edu.reset_index()
def percentToFloat(percent):
    return float(percent.strip('%'))/100
immigration_edu['9th-12th Grade '] = immigration_edu['9th-12th Grade '].map(percentToFloat)
immigration_edu['High School Diploma or Equivalent'] = immigration_edu['High School Diploma or Equivalent'].map(percentToFloat)
immigration_edu['Some College or Associate\'s Degree'] = immigration_edu['Some College or Associate\'s Degree'].map(percentToFloat)
immigration_edu['Less than 9th Grade'] = immigration_edu['Less than 9th Grade'].map(percentToFloat)
immigration_edu['Bachelor\'s Degree or Higher'] = immigration_edu['Bachelor\'s Degree or Higher'].map(percentToFloat)
immigration_edu.head(5)
```

```python
immigration_edu.to_csv()
```
